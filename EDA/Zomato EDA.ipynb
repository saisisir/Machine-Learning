```python
#find row counts of a group of columns , reset index and later rename them
ratings = df2.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'nrows'})
```

```python
ratings
```

#Observation: 
    # 1. when rating is between 4.5 to 4.9 --->>> Excellent
    # 2. when rating is between 4.0 to 4.4 --->>> Very Good
    # 3. when rating is between 3.5 to 3.9 --->>> Good
    # 4. when rating is between 2.5 to 3.4 --->>> Average
    # 5. when rating is between 2.0 to 2.2 --->>> Poor
    # 6. when rating is 0.0                --->>> Most people have not rated yet

```python
import matplotlib
#organizing the plots by using figure and figsize as rcParams in matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)

#Visualizing plots in seaborn, can also use palette
sns.barplot(x = 'Aggregate rating', y = 'nrows', hue = 'Rating color', data = ratings, 
            palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])   
```

```python
# Observation: 
#     1. No rating count is very high
#     2. Max ratings are between 2.6 to 4.3

#Count plots/Frequency in Seaborn
sns.countplot(x = 'Rating text', data = ratings)
```
