```python
#appending both dataframes into a final one 
data = pd.concat([train, test]).reset_index(drop = True)
```

```python
#Splitting day into date, month and year
data[['day', 'month', 'year']] = data['Date_of_Journey'].str.split('/', expand = True)
```

```python
data.head()
```

```python
data.day = data.day.astype(int)
data.month = data.month.astype(int)
data.year = data.year.astype(int)
```

```python
data.info()
```

```python
data = data.drop('Date_of_Journey', axis = 1)
```

```python
data.Arrival_Time.str.split(' ', expand = True)[1]
```

```python
data.Arrival_Time.str.split(' ', expand = True)[1]
```

```python
#Split timewith date to time, day and month
data['Arrival_Time1'] = data.Arrival_Time.str.split(' ', expand = True)[0]
data['Arrival_Day'] = data.Arrival_Time.str.split(' ', expand = True)[1]
data['Arrival_Month'] = data.Arrival_Time.str.split(' ', expand = True)[2]
data['Arrival_Hour'] = data.Arrival_Time1.str.split(':', expand = True)[0]
data['Arrival_Min'] = data.Arrival_Time1.str.split(':', expand = True)[1]
```

```python
#dropping both time columns after extracting necessary information
data.drop(['Arrival_Time1', 'Arrival_Time'], axis = 1, inplace = True)
```

```python
#converting strings into integers
data.Arrival_Hour = data.Arrival_Hour.astype(int)
data.Arrival_Min = data.Arrival_Min.astype(int)
```

```python
#splitting departure time to hour and minutes
data['Dep_Hour'] = data.Dep_Time.str.split(':', expand = True)[0]
data['Dep_Min'] = data.Dep_Time.str.split(':', expand = True)[1]
```

```python
data.drop(['Dep_Time'], axis = 1, inplace = True)

data.Dep_Hour = data.Dep_Hour.astype(int)
data.Dep_Min = data.Dep_Min.astype(int)
```

```python
data.info()
```

```python
data['Airline'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder  #(to be used only for input features)
from sklearn.preprocessing import LabelEncoder #(to be used only for target variables)

data['airline_enc'] = OrdinalEncoder().fit_transform(data['Airline'].values.reshape(-1, 1))
```

```python
data.head()
```

```python
data['airline_enc1'] = LabelEncoder().fit_transform(data['Airline'])
```

```python
data.head()
```

```python
data.drop(['Airline', 'airline_enc1'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
#Change New Delhi to delhi to have a common name across source and destination
data['Destination'] = data.Destination.str.replace('New Delhi', 'Delhi')
```

```python
data.head()
```

```python
#Use ordinalEncoder to convert independent features into integers
data[['source', 'destn']] = OrdinalEncoder().fit_transform(data[['Source', 'Destination']].values.reshape(-1, 2))
```

```python
#Drop Multiple columns at once
data.drop(['Source', 'Destination', 'Route'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.Total_Stops.unique()
```

```python
#visulaize which row has NaN in Total_Stops column
data[data.Total_Stops.isnull()]
```

```python
#Map and change categorical data into number 
data['total_stops'] = data['Total_Stops'].map({'non-stop':0, '2 stops': 2, '1 stop': 1, '3 stops': 3, '4 stops': 4, 'NaN':1})
```

```python
data.total_stops.unique()
```

```python
#Splitting duration into hour and minutes
data['duration_hour'] = data.Duration.str.split(' ', expand = True)[0]
data['duration_min'] = data.Duration.str.split(' ', expand = True)[1]
```

```python
data.head()
```

```python
#splitting text portion from hour and minutes
data['duration_hour'] = data.duration_hour.str.split('h', expand = True)[0]
data['duration_min'] = data.duration_min.str.split('m', expand = True)[0]
```

```python
#sometimes, None is equivalent of null and can be replaced using fillna
data['duration_min'] = data.duration_min.fillna(0)
data['total_stops'] = data['total_stops'].fillna(0)
```

```python
data.total_stops.isnull().sum()
```

```python
data.drop('Duration', axis = 1, inplace = True)
```

```python
#Replace incorrect hour values into decimals
data['duration_hour'].unique()
```

```python
#Replace incorrect hour values into decimals. We see a value of 5m
data['duration_hour'] = data['duration_hour'].replace('5m', 0.12)
```

```python
data['duration_hour'] = data['duration_hour'].astype(int)
data['duration_min'] = data['duration_min'].astype(int)
```

```python
data.head()
```

```python
#Converting hour and minutes into minutes
data['total_duration_mins'] = data['duration_hour']* 60 + data['duration_min']
```

```python
data.head()
```

```python
#drop temp hour and min columns
data.drop(['duration_hour', 'duration_min'], axis = 1, inplace = True)
```

```python
data.Additional_Info.unique()
```

```python
#change format of two similar but diff 'no info' to one 
data['Additional_Info'] = data.Additional_Info.str.replace('No info', 'No Info')
```

```python
#Convert strings into unique codes using ordinalencoder
data.Additional_Info = OrdinalEncoder().fit_transform(data['Additional_Info'].values.reshape(-1, 1))
```
