```python
#If weekend has to be excluded
dates = pd.date_range(start = '2017-06-01', end = '2017-06-30', freq = 'B')
```

```python
data2.head()
```

```python
#to include weekends carrying over fridays stock price during close
data2.asfreq(freq ='D', method = 'pad')
```

```python
#to include hourly data 
data2.asfreq(freq ='H', method = 'pad')
```

```python
#to show weekly data 
data2.asfreq(freq ='W', method = 'pad')
```
