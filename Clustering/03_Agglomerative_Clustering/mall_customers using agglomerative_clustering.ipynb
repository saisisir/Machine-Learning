```python
df = df.iloc[:, 3:5].values
```

```python
#PLot dendrogram
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(12, 6))
plt.title('Customer Dendrograms')
dend = sch.dendrogram(sch.linkage(df, method = 'ward'))
```

```python
#The number of vertical lines tells us how many clusters should they be grouped into
                    
#import agglomerative clustering
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5)
y_means = cluster.fit_predict(df)
y_means
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df[:,0], df[:,1], c=cluster.labels_)

from sklearn import metrics
print(metrics.silhouette_score(df, y_means))
```
