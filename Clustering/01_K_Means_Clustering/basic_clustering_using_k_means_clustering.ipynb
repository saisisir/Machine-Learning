```python
#Scatterplot
plt.scatter(data['x'], data['y'])
sns.scatterplot(data = data)
```

```python
#import kmeans
from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(data)
    wcss.append(km.inertia_)
    
wcss
```

```python
plt.plot(range(1,11), wcss)
# sns.scatterplot(range(1,11), wcss)
plt.title("The Elbow Method")
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Elbow method hints at 2

x = data.iloc[:, :].values
km = KMeans(n_clusters = 2)
y_means = km.fit_predict(x)
```

```python
#All clusters coming in 0 cluster
x[y_means == 0]

#All points in cluster 1
x[y_means == 1]

#All first indexes of points in cluster 0
x[y_means == 0, 0]
```

```python
#Plotting final scatter plot
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 10, c = 'red', label = 'Cluster 0')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 10, c = 'blue', label = 'Cluster 1')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 10, c = 'magenta')
plt.title('x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
